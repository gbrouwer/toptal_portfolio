### Create Blobs

```python
x, y = np.mgrid[-1:1:.01, -1:1:.01]
pos = np.dstack((x, y))
rv = mnorm([0.5, -0.2], [[2.0, 0.3], [0.3, 0.5]])
fig2 = plt.figure(figsize=(12,12))
ax2 = fig2.add_subplot(111)
im = rv.pdf(pos)
plt.imshow(im)
```
