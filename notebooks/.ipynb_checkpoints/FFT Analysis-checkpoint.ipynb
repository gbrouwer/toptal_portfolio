### Import

```python
from scipy.fft import fft
from scipy.fft import ifft
from scipy.fft import fftfreq
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
```

### Nomimal Signal

```python
N = 60 * 30
T = 1 / N
x = np.linspace(0.0, N*T, N, endpoint=False)
y1 = np.sin(14.0 * 2.0*np.pi*x)
y2 = np.sin(14.0 * 2.0*np.pi*x + 0.25)
n1 = np.random.randn(len(y1)) * 0.1
n2 = np.random.randn(len(y1)) * 0.1

y1 = y1 + n1
y2 = y2 + n2

yf1 = fft(y1)
xf1 = fftfreq(N, T)[:N//2]
yf1 = 2.0/N * np.abs(yf1[0:N//2])
xf1 = xf1[:50]
yf1 = yf1[:50]

yf2 = fft(y2)
xf2 = fftfreq(N, T)[:N//2]
yf2 = 2.0/N * np.abs(yf2[0:N//2])
xf2 = xf2[:50]
yf2 = yf2[:50]


plt.figure(figsize=(10,5))
plt.plot(xf1,yf1,'k')
plt.plot(xf2,yf2,'r')
plt.xlabel('Frequency',fontsize=18)
plt.ylabel('Power',fontsize=18)
plt.grid()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(y1,'k')
plt.plot(y2,'r')
plt.xlabel('Time',fontsize=18)
plt.ylabel('Relative Displacement',fontsize=18)
plt.grid()
plt.show()
```
