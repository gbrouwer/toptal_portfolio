### Create Color Wheel

```python
plt.figure(figsize=(8,8))
val = np.linspace(0,2*np.pi,33)[:-1]
x = np.sin(val) * 60
y = np.cos(val) * 60
LAB = np.zeros((32,3))
LAB[:,0] = 60
LAB[:,1] = x
LAB[:,2] = y
RGB = color.lab2rgb(LAB)
plt.scatter(x,y,c=RGB,s=500)
plt.savefig('../assets/neuroscience/color_categories.png',dpi=300)
```
