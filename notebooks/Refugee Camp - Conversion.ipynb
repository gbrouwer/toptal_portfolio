### Create Color Wheel

```python
im = Image.open('../assets/settlement_detection/ztari_1.png') 
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im)
```

```python
LAB = color.rgb2lab(im)
```

```python
saturation = np.sqrt(LAB[:,:,1]**2 + LAB[:,:,2]**2)
saturation = saturation**0.25
saturation = saturation - np.min(saturation)
saturation = saturation / np.max(saturation)
saturation = 1 - saturation
plt.figure(figsize=(10,10))
plt.imshow(saturation,cmap='gray')
saturation = saturation * 255
saturation = saturation.astype('uint8')
saturation = Image.fromarray(saturation) 

saturation.save('../assets/settlement_detection/ztara2.png')
```
